# Processing Data

# Rename Columns

```python
data.rename(columns = {"1. What is your age?":"Age"}, inplace = True)
data.rename(columns = {"2. Gender":"Gender"}, inplace = True)
data.rename(columns = {"3. Relationship Status":"Relationship Status"}, inplace = True)
data.rename(columns = {"4. Occupation Status":"Occupation Status"}, inplace = True)
data.rename(columns = {"5. What type of organizations are you affiliated with?":"Organization Type"}, inplace = True)
data.rename(columns = {"6. Do you use social media?":"Social Media Status"}, inplace = True)
data.rename(columns = {"7. What social media platforms do you commonly use?":"Social-Media Used"}, inplace = True)
data.rename(columns = {"8. What is the average time you spend on social media every day?":"Average_Time"}, inplace = True)
data.rename(columns = {"9. How often do you find yourself using Social media without a specific purpose?":"Used without purpose"}, inplace = True)
data.rename(columns = {"10. How often do you get distracted by Social media when you are busy doing something?":"Distraction"}, inplace = True)
data.rename(columns = {"11. Do you feel restless if you haven't used Social media in a while?":"Non Social-Media Restlessness"}, inplace = True)
data.rename(columns = {"12. On a scale of 1 to 5, how easily distracted are you?":"Easily Distracted"}, inplace = True)
data.rename(columns = {" Easily Distracted Rate":"Easily Distracted"}, inplace = True)
data.rename(columns = {"13. On a scale of 1 to 5, how much are you bothered by worries?":"Worries' Impact"}, inplace = True)
data.rename(columns = {"14. Do you find it difficult to concentrate on things?":"Difficulty Concentrating"}, inplace = True)
data.rename(columns = {"15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?":"Social Media Comparison"}, inplace = True)
data.rename(columns = {"16. Following the previous question, how do you feel about these comparisons, generally speaking?":"Comparative Feelings"}, inplace = True)
data.rename(columns = {"17. How often do you look to seek validation from features of social media?":"Social-Media Validation "}, inplace = True)
data.rename(columns = {"18. How often do you feel depressed or down?":"Depressive Frequency"}, inplace = True)
data.rename(columns = {"19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?":"Daily Activity Fluctuation"}, inplace = True)
data.rename(columns = {"20. On a scale of 1 to 5, how often do you face issues regarding sleep?":"Sleep Issue"}, inplace = True)
```
